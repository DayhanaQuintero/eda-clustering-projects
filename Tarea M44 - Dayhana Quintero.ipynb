# Feature Engineering

```python
#Eliminamos las columnas que no aportan informacion relevante 
df = df.drop(columns=["EmployeeID", "EmployeeCount", "Over18", "StandardHours"])
```

```python
#Creación de nueva columna
#Ingreso por nivel de puesto
df["IncomePerJobLevel"] = df["MonthlyIncome"] / df["JobLevel"]
```

```python
df.head(5)
```

```python
#Conversion de Attrition a variable numérica
df["Attrition"] = df["Attrition"].map({"No": 0, "Yes": 1})
```

```python
df.head(3)
```

```python
df.select_dtypes(include="object").columns
```

```python
#Creación de dummies
df = pd.get_dummies(df, drop_first=True)
```

```python
#Separamos X y Y
X = df.drop(columns=["Attrition"])
y = df["Attrition"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
#Dividir en entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1, stratify=y)
```

```python
#Estandarización
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
regression_model = LogisticRegression(solver="liblinear", class_weight="balanced")
regression_model.fit(x_train_scaled, y_train)
```

```python
predictions = regression_model.predict(x_test_scaled)
predictions
```

```python
#Probabilidades de prediction
predictions_prob = regression_model.predict_proba(x_test_scaled)[:,1]
predictions_prob[:10]
```

```python
#Medidas del modelo
print("Medidas del Modelo")
print("--------------------")

print("Accuracy =", accuracy_score(y_test, predictions))
print("Precision =", precision_score(y_test, predictions))
print("Recall =", recall_score(y_test, predictions))
print("F1 Score =", f1_score(y_test, predictions))
print("Roc Auc Score =", roc_auc_score(y_test, predictions_prob))
```

```python
#Matriz de confusión
cm = confusion_matrix(y_test, predictions)
cm
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds")
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.title("Matriz de Confusión")
plt.show()
```

```python
#Análisis de sensibilidad
coeff_data = pd.DataFrame({"Variable": x_train.columns,"Coeficiente": regression_model.coef_[0]})

coeff_data.sort_values(by="Coeficiente", ascending=False).head(10)
```

```python
coeff_data.sort_values(by="Coeficiente", ascending=True).head(10)
```

```python
#Importancia de variables
top_features = coeff_data.reindex(coeff_data["Coeficiente"].abs().sort_values(ascending=False).index).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top_features, x="Coeficiente", y="Variable")
plt.title("Variables más influyentes en Attrition")
plt.show()
```

```python
#Probabilidades predichas
plt.figure(figsize=(8,4))
sns.histplot(predictions_prob, bins=20, kde=True)
plt.title("Distribución de probabilidades predichas")
plt.xlabel("Probabilidad de Attrition")
plt.show()
```
